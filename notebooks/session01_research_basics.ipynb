## 0. なぜ論文（客観的証拠）が必要なのか？

私たちが新しい発見をしたり、自分の意見を他人に伝えようとするとき、単に「私はこう思う」「私の経験ではこうだった」と言うだけでは、相手を納得させることは困難です。

客観的な証拠（エビデンス）がない主張は、第三者からは**「単なる個人の思い込み」**や**「偶然の出来事」**と見なされてしまう可能性があります。

### 証拠（論文）の役割
1. **客観性の確保**: 誰が見ても納得できるデータや分析結果を示すことで、独りよがりな解釈を防ぎます。
2. **再現性の保証**: どのような手順でその結果を得たかを公開することで、他人が同じ方法で確認できるようにします。
3. **知識の共有と蓄積**: 正しいと検証された知識を論文として残すことで、人類全体の知的な財産として積み上がっていきます。

特に医学やデータサイエンスの分野では、**「人の命に関わる意思決定」**を扱うため、直感や思い込みではなく、厳密に検証された「証拠」に基づく判断（EBM: Evidence-Based Medicine）が不可欠です。

## 1. 研究とは何か？

### 1.1 研究の定義

**研究（Research）**とは：
- 新しい知識を発見すること
- 既存の知識を検証・拡張すること
- 問題を解決するための体系的な調査

### 1.2 研究のプロセス

```
1. 問題の発見・定式化
   ↓
2. 文献調査（先行研究の確認）
   ↓
3. 仮説の設定
   ↓
4. 研究デザインの設計
   ↓
5. データ収集
   ↓
6. データ分析
   ↓
7. 結果の解釈
   ↓
8. 論文執筆・発表
```

### 1.3 良い研究の条件

- **新規性（Novelty）**: 新しい発見や視点がある
- **妥当性（Validity）**: 方法論が適切で、結果が信頼できる
- **再現性（Reproducibility）**: 他の研究者が同じ結果を得られる
- **意義（Significance）**: 学術的・社会的に価値がある

## 2. 論文の構造：IMRAD

### 2.1 IMRADとは？

科学論文の標準的な構成：

- **I**ntroduction（序論）
- **M**ethods（方法）
- **R**esults（結果）
- **A**nd
- **D**iscussion（考察）

---

### 2.2 【実例】アブストラクトで見るIMRAD

アブストラクト（要旨）は、論文全体のIMRAD構造を短くまとめたものです。多くの場合、アブストラクト自体も構造化されています。

**例：放射線科専門医試験におけるLLMの性能評価（架空の抄録）**

> **[Introduction]** 近年、大規模言語モデル（LLM）の医学的知識が注目されているが、日本の放射線科専門医試験における性能は不明である。本研究の目的は、最新のLLMがこの試験で合格ラインに達するかを検証することである。
>
> **[Methods]** 2023年度の放射線科専門医試験の筆記問題（100問）を対象とした。GPT-4およびClaude 3に対し、画像情報を除いたテキスト問題を入力し、正答率を算出した。
>
> **[Results]** GPT-4の正答率は85%（85/100）、Claude 3の正答率は78%（78/100）であった。いずれも実際の合格ラインである70%を大きく上回った。特に物理学や解剖学の分野で高い正答率を示した。
>
> **[Discussion]** 最新のLLMは日本の放射線科専門医レベルの知識を十分に有していることが示唆された。しかし、画像診断の核心である画像解釈能力の検証が今後の課題である。

---

### 2.3 各セクションの役割

#### Introduction（序論）
**目的**: 研究の背景と目的を説明

含まれる内容：
- 研究分野の背景
- 先行研究のレビュー
- 研究の必要性（ギャップ）
- 研究の目的・仮説

**読者への問い**: 「なぜこの研究が必要なのか？」

> **実例文**: 近年、大規模言語モデル（LLM）の医学的知識が注目されているが、日本の放射線科専門医試験における性能は不明である。本研究の目的は、最新のLLMがこの試験で合格ラインに達するかを検証することである。


#### Methods（方法）
**目的**: 研究の手法を詳細に記述

含まれる内容：
- 研究デザイン
- データ収集方法
- サンプルサイズ、対象者
- 分析手法（統計手法、機械学習モデル）
- 使用したソフトウェア・ツール

**読者への問い**: 「どうやって研究したのか？」

> **実例文**: 2023年度の放射線科専門医試験の筆記問題（100問）を対象とした。GPT-4およびClaude 3に対し、画像情報を除いたテキスト問題を入力し、正答率を算出した。


#### Results（結果）
**目的**: 分析結果を客観的に報告

含まれる内容：
- 記述統計（平均、標準偏差など）
- 統計的検定の結果（p値、信頼区間）
- 図表（グラフ、表）
- 機械学習モデルの性能（精度、AUCなど）

**読者への問い**: 「何がわかったのか？」

> **実例文**: GPT-4の正答率は85%（85/100）、Claude 3の正答率は78%（78/100）であった。いずれも実際の合格ラインである70%を大きく上回った。特に物理学や解剖学の分野で高い正答率を示した。

**注意**: 結果の解釈はここでは行わない（Discussionで行う）

#### Discussion（考察）
**目的**: 結果の意味を解釈し、研究の意義を論じる

含まれる内容：
- 結果の解釈
- 先行研究との比較
- 研究の限界（Limitations）
- 今後の研究方向
- 結論（Conclusion）

**読者への問い**: 「結果は何を意味するのか？」

> **実例文**: 最新のLLMは日本の放射線科専門医レベルの知識を十分に有していることが示唆された。しかし、画像診断の核心である画像解釈能力の検証が今後の課題である。


### 2.4 その他のセクション

- **Abstract（要旨）**: 論文全体の要約（200-300語）
- **Keywords（キーワード）**: 論文の内容を表す重要な単語（3-5個程度）。検索の目印になる
- **Abbreviations（略語一覧）**: 論文内で使用される略語の定義
- **References（参考文献）**: 引用した文献のリスト
- **Acknowledgments（謝辞）**: 資金提供者や協力者への感謝
- **Supplementary Materials（補足資料）**: 本文に載せきれない詳細なデータや図表

## 3. なぜ統計学・機械学習が必要か？

### 3.1 データサイエンス研究の特徴

現代の研究では、大量のデータを扱うことが一般的：
- 医学：電子カルテ、医学画像、ゲノムデータ
- 社会科学：アンケート、SNSデータ
- 工学：センサーデータ、シミュレーション結果

→ **統計学・機械学習が不可欠**

### 3.2 統計学の役割

#### 記述統計
- データの特徴を要約（平均、分散、分布）
- 可視化（ヒストグラム、散布図）

#### 推論統計
- サンプルから母集団の性質を推定
- 仮説検定（差があるか？関連があるか？）
- 信頼区間の計算

**母集団と標本**:  
研究の対象となるグループ全体を**母集団（Population）**と呼びますが、現実的にその全員を調査することは不可能です。そのため、一部のデータ（**標本：Sample**）を抽出し、そこから全体の特徴を推測します。

**推論の流れ**:
1. **サンプリング**: 母集団から一部のデータを集める
2. **統計的推論**: 集めたデータの結果に基づき、母集団全体でも同じことが言えるかを確率的に判断する

**例**: 新薬の効果を検証
- 仮説：「新薬は既存薬より効果がある」
- 方法：臨床試験でデータ収集
- 分析：t検定でp値を計算
- 結論：p < 0.05なら「統計的に有意な差がある」

### 3.3 機械学習の役割

#### 予測モデルの構築
- 過去のデータから未来を予測
- 例：患者の予後予測、疾患の診断

#### パターンの発見
- データから隠れたパターンを抽出
- 例：遺伝子発現データのクラスタリング

#### 自動化
- 人間が行っていた作業を自動化
- 例：医学画像の自動診断

### 3.4 統計学 vs 機械学習

| 項目     | 統計学                   | 機械学習                                   |
| -------- | ------------------------ | ------------------------------------------ |
| 目的     | 仮説検定、因果推論       | 予測、パターン認識                         |
| 解釈性   | 高い（係数の意味が明確） | 低い（ブラックボックス）                   |
| データ量 | 少量でもOK               | 大量のデータが必要                         |
| 例       | t検定、回帰分析          | ニューラルネットワーク、ランダムフォレスト |

**重要**: どちらか一方ではなく、**両方を使い分ける**ことが重要

## 4. 医学データサイエンス研究の例

### 4.1 新型コロナウイルスのマスク・遮蔽板関連

「新型コロナウイルスの空気伝播に対するマスクの防御効果」

https://www.ims.u-tokyo.ac.jp/imsut/jp/about/press/page_00042.html

元になった英語論文
Effectiveness of Face Masks in Preventing Airborne Transmission of SARS-CoV-2

https://journals.asm.org/doi/full/10.1128/msphere.00637-20

-> マスクは感染予防効果に重要


新型コロナウイルス感染対策による聴き取り阻害を定量化する試み

https://www.jstage.jst.go.jp/article/audiology/66/1/66_56/_article/-char/ja/

-> 遮蔽板とマスクを併用した場合は聞き取り阻害の原因になりうる


### 4.2 論文は対象と目的・結論をセットで考えることが重要

上記のような論文ではマスクは良いものとしても悪いものとしても報告されている。

しかし、これらの論文は**矛盾している**わけではありません。それぞれの論文が**「誰を対象に」「何を目的として」「どのような条件で」**研究を行ったかが異なるからです。

|                        | 論文①                                      | 論文②                                              |
| ---------------------- | ------------------------------------------ | -------------------------------------------------- |
| **対象（Population）** | COVID-19への感染リスクがある一般集団       | 聴覚コミュニケーションを行う環境（学校・会議など） |
| **目的（Objective）**  | マスクが飛沫・エアロゾルの感染拡大を防ぐか | マスク＋遮蔽板が聴き取りに与える影響               |
| **結論（Conclusion）** | マスクは感染予防に有効である               | 遮蔽板とマスクの併用は聴き取りを阻害しうる         |

#### ポイント：「誰に・何について・どのような条件で」という文脈が重要

論文の結論だけを切り取ると、「マスクは良い」「マスクは悪い」という**正反対の主張**に見えます。

しかし実際には、両方の論文とも**それぞれの研究文脈では正しい**です。

> ❌ **NG**: 「マスクは効果があるという論文がある → だからマスクは絶対に正しい」  
> ✅ **OK**: 「この論文は○○を対象に、△△という条件で、□□を測定した結果、マスクに感染予防効果が認められた」

論文を読む際には、以下の点を必ず確認しましょう：

1. **対象（Population）**: どんな集団・環境を研究しているか？
2. **目的（Objective/Intervention）**: 何を調べているか？
3. **結論（Conclusion）**: その条件のもとで何が言えるか？（限界は？）

この視点は、後に学ぶ**PICO/PECO**（研究を整理するフレームワーク）の基礎にもなります。

## 🐍 Python 基礎構文まとめ

データサイエンスを実践するうえで必要な Python の基本を整理します。

### 0. なぜデータサイエンスで「遅い」Pythonが使われるのか？

Pythonは他のプログラミング言語と比較して、**実行速度が極端に遅い**という弱点があります。それにも関わらず、なぜデータサイエンスの標準言語となっているのでしょうか？

**① 実際の重い処理は「高速な言語」で行われている**  
Pythonで利用するデータ分析ライブラリ（PandasやNumPy）や機械学習ライブラリの内部は、実はC言語やC++といった非常に実行速度の速い言語で作られています。Python自体はそれらを指示する「操作パネル」として機能しているため、計算などの実作業では高速な処理が可能になっています。

**② 現場での「スクラップ・アンド・ビルド」のしやすさ**  
データ分析の現場では、最初から完璧なシステムを作ることはありません。データを見て仮説を立て、コードを書いて試し、直すという「スクラップ・アンド・ビルド（試行錯誤）」をひたすら繰り返します。Pythonはコードが直感的で短く書けるため、この「アイデアを素早く試す」スピードが他の言語より圧倒的に優れています。

**⚠️ Pythonの欠点：バージョン・環境管理の難しさ**  
Pythonの厄介な問題の一つが「環境の構築」と「バージョン管理」です。様々なライブラリを組み合わせて使うため、「ライブラリのバージョンが少し違うだけでコードが動かなくなる」「自分のPCでは動くが他人のPCでは動かない（環境依存）」といったトラブルが日常茶飯事です。本ゼミで各自のPCにPythonを直接インストールせず、Google Colabを使用しているのもこの問題を回避するためです。

### 3. 文字列操作と f-string
データのラベル処理などで文字列操作は欠かせません。Pythonの文字列は非常に柔軟で、簡単な構文で様々な操作が可能です。特に **f-string** は、変数の値や計算結果を文章の中に埋め込む際に非常に便利で、データ分析結果の出力などによく使われます。


````python
s = "Data Science"

# 基本的な操作
print(len(s))                    # len(): 文字数を取得（スペースも1文字としてカウント）
print(s.upper())                 # .upper(): 全て大文字に変換 -> "DATA SCIENCE"
print(s.lower())                 # .lower(): 全て小文字に変換 -> "data science"
print(s.replace("Science", "Analysis")) # .replace(): 文字列の置換 -> "Data Analysis"

# スライス（部分文字列の抽出）
# [開始位置:終了位置] で指定します。開始位置は0から数え、終了位置の文字は含まれません。
print(s[0:4])                    # 最初から4文字分を取得 -> "Data"
print(s[-7:])                    # 後ろから7文字分から最後まで取得 -> "Science"

# f-string（フォーマット文字列）
# 文字列の前に f を付け、{ } の中に変数名や式を直接書きます。
name = "Alice"
age = 25
x = 3.14159

print(f"私の名前は{name}で、年齢は{age}歳です。")
print(f"円周率は {x:.2f} です")  # :.2f をつけると「小数点2桁まで」丸めて表示
print(f"5年後は {age + 5} 歳になります。") # { } の中で直接計算も可能
````

### 4. リストとタプル
複数のデータをまとめて扱うためのデータ構造です。データ分析では、複数の患者データや時系列データをまとめて処理するために頻繁に使用します。

- **リスト (List)**: `[ ]` で囲みます。中身を後から自由に変更（追加・更新・削除）できるのが特徴です。
- **タプル (Tuple)**: `( )` で囲みます。一度作成すると中身を変更できません（イミュータブル）。座標や設定値など、予期せず変更されたくないデータに用います。


````python
# --- リストの操作 ---
scores = [85, 78, 92, 61, 73]

# 要素の取得（インデックスは 0 から始まります）
print(scores[0])    # 最初の要素 -> 85
print(scores[-1])   # 最後の要素（-1で一番後ろを指定できる） -> 73

# リストの変更
scores[1] = 80      # 2番目の要素（78）を80に変更
scores.append(88)   # .append(): 末尾に新しいデータを追加
print(scores)       # [85, 80, 92, 61, 73, 88]

# リストの便利なメソッド
scores.sort()           # .sort(): 昇順（小さい順）に並び替え
scores.sort(reverse=True) # 降順（大きい順）に並び替え
length = len(scores)    # len(): 要素数（データの個数）を取得

# --- タプルの操作 ---
coordinates = (35.6895, 139.6917) # 東京の緯度経度
print(coordinates[0])             # タプルもリストと同じように取得可能

# coordinates[0] = 36.0           # ←中身を変更しようとするとエラー（TypeError）になる
````

### 6. 条件分岐 (if / elif / else)
「もし〜ならA、そうでなければB」という処理の流れを作ります。判定の基準となる条件式（比較演算子：`==`, `!=`, `<`, `>`, `<=`, `>=` など）を組み合わせて使います。
Pythonでは、**インデント（字下げ）**を使ってブロック（処理のまとまり）を表現する点に注意してください。


````python
bmi = 27.0

# 基本的な if / elif / else 構文
if bmi < 18.5:
    # bmiが18.5未満の場合に実行される
    print("低体重")
elif bmi < 25.0:
    # 上の条件を満たさず、bmiが25.0未満の場合に実行される
    print("普通体重")
elif bmi < 30.0:
    # さらに上の条件を満たさず、bmiが30.0未満の場合
    print("過体重")
else:
    # 上記のすべての条件を満たさなかった場合に実行される
    print("肥満")

# 複数の条件を組み合わせる (and, or, not)
age = 45
is_smoker = True

if age >= 40 and is_smoker:
    print("特定健診と、喫煙によるリスク評価を推奨します。")
elif age >= 40 or is_smoker:
    # どちらか一方だけ当てはまる場合
    print("定期的な健康診断を推奨します。")
````

### 7. 繰り返し (for / while)
配列内のすべてのデータに同じ計算を適用したり、特定の条件を満たすまで処理を続けるなど、大量のデータ処理で非常に重要になります。データ分析で最もよく使うのはリストなどを対象とした `for` ループです。


````python
# --- for ループ ---
# リストの中から要素を一つずつ変数に取り出して繰り返す
patients = ["田中", "鈴木", "佐藤"]
for p in patients:
    print(f"患者名: {p} さん")

# range() を使った指定回数の繰り返し
# range(5) は 0, 1, 2, 3, 4 を生成します（5は含まれないことに注意）
for i in range(3):
    print(f"{i+1}回目の処理です")

# enumerate() を使うと、インデックス（番号）と要素を同時に取得できます
for i, p in enumerate(patients):
    print(f"ID{i+1}: {p} さん")

# --- while ループ ---
# 条件が True である限り繰り返す
count = 0
while count < 3:
    print(f"現在のカウント {count}")
    count += 1  # count = count + 1 と同じ。これを忘れると無限ループになるので注意

# ループの制御 (break, continue)
# break: ループを途中で完全に終了する
# continue: 今回のループ処理をそこで打ち切り、次のステップへスキップする
````

### 8. 関数 (def)
何度も使う処理をひとまとめにして名前を付けたものです。自分で関数を作ることで、コードの重複を防ぎ、見やすくエラーの少ないプログラムを書くことができます。
関数は「入力（引数：ひきすう）」を受け取り、処理を行った後、「結果（戻り値：もどりち）」を返します。


````python
# 1. 関数の定義
def calc_bmi(weight_kg, height_m):
    """
    BMIを計算する関数 (ここはドックストリングと呼ばれ、関数の説明を書きます)
    """
    bmi = weight_kg / (height_m ** 2)
    return bmi  # return で計算結果を呼び出し元に返す

# 2. 関数の呼び出し
# 引数として 65 と 1.70 を渡すと、計算結果が my_bmi に代入される
my_bmi = calc_bmi(65, 1.70)
print(f"BMIは {my_bmi:.1f} です")

# 3. 複数の戻り値を返す関数
def get_stats(data_list):
    """リストの最小値、最大値、平均値を計算してまとめて返す関数"""
    minimum = min(data_list)
    maximum = max(data_list)
    average = sum(data_list) / len(data_list)

    # 複数の変数をカンマ区切りで return すると、タプルとして返される
    return minimum, maximum, average

# 複数の戻り値を受け取る（アンパック代入）
stats_data = [120, 110, 135, 128]
min_val, max_val, avg_val = get_stats(stats_data)
print(f"最小: {min_val}, 最大: {max_val}, 平均: {avg_val}")
````

### 9. リスト内包表記 (List Comprehension, ちょっと高度ですが便利です)
Python特有の非常に便利な書き方です。「既存のリストから新しいリストを作る」という処理を、1行で簡潔に記述できます。通常の `for` ループを使うよりも実行速度が速いため、データの前処理などで頻繁に活用されます。


````python
# 通常の for ループでリストを作る場合
squares_normal = []
for x in range(1, 6):
    squares_normal.append(x ** 2)
print("通常:", squares_normal)

# --- 上記をリスト内包表記で書く場合 ---
# 構文: [適用したい式 for 変数 in 取り出す対象のリストなど]
squares = [x ** 2 for x in range(1, 6)]
print("内包表記:", squares)

# さらに条件で絞り込む場合
# 構文: [適用したい式 for 変数 in リストなど if 条件式]
numbers = [10, 15, 20, 25, 30]

# 20以上の数字だけをそのまま取り出す
large_nums = [x for x in numbers if x >= 20]
print("20以上:", large_nums)

# 20以上の数字を半分にして取り出す
halved_large_nums = [x / 2 for x in numbers if x >= 20]
print("20以上を半分に:", halved_large_nums)
````
