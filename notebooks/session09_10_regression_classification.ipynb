## 4. 標準化のポイント

標準化とは、特徴量を以下のように変換する処理です。

```
標準化後の値 = (元の値 - 平均) / 標準偏差
```

標準化後は、各特徴量の平均が約0、標準偏差が約1になります。  
SVM、Logistic回帰、k近傍法、Neural Network のようなモデルでは、特徴量のスケールが性能に影響しやすいため、標準化が重要です。

一方で、決定木やRandom Forestは特徴量の大小関係を使って分岐するため、標準化の影響は比較的小さいです。

> [!IMPORTANT]
> テストデータの情報が学習に漏れないように、実際のモデル学習では `Pipeline` の中で標準化します。

## 5. ハイパーパラメータチューニング

ハイパーパラメータとは、モデルがデータから自動で学習する値ではなく、人間が事前に決める設定値のことです。

たとえば、SVMの `C` や `gamma`、Random Forestの `n_estimators` や `max_depth` はハイパーパラメータです。これらの値を変えると、モデルの複雑さや予測性能が変わることがあります。

### 5-1. チューニングで注意すること

ハイパーパラメータチューニングでは、**テストデータを使って最良パラメータを選んではいけません**。テストデータは、最後にモデルの性能を確認するために残しておくデータです。

今回の流れは以下です。

1. 訓練データとテストデータに分ける
2. 訓練データの中だけで交差検証を行う
3. `GridSearchCV` で良いハイパーパラメータを探す
4. 最良モデルをテストデータで評価する

`GridSearchCV` は、指定したハイパーパラメータの候補をすべて試し、交差検証の平均スコアが最も高い組み合わせを選びます。

今回は評価指標として `roc_auc` を使います。`roc_auc` は、分類の閾値を変えたときの性能をまとめて評価できる指標です。

### 5-2. SVMで調整する主なハイパーパラメータ

SVMでは、特に以下が重要です。

| パラメータ | 意味 |
| ---- | ---- |
| `C` | 誤分類をどの程度許すか |
| `gamma` | 1つのデータ点が分類境界に与える影響の広さ |
| `kernel` | 分類境界の形 |

`C` が大きいほど、訓練データをできるだけ正確に分類しようとします。ただし、大きすぎると訓練データに合わせすぎることがあります。

`gamma` が大きいほど、1つ1つのデータ点の影響範囲が狭くなり、複雑な分類境界を作りやすくなります。ただし、大きすぎると過学習しやすくなります。

### 5-7. Random Forestで調整する主なハイパーパラメータ

Random Forest は、決定木をたくさん作り、それらの多数決で分類するモデルです。標準化の影響は比較的小さいですが、ハイパーパラメータによって性能や過学習のしやすさが変わります。

| パラメータ | 意味 |
| ---- | ---- |
| `n_estimators` | 作成する決定木の本数 |
| `max_depth` | それぞれの決定木の最大の深さ |
| `min_samples_split` | ノードを分割するために必要な最小サンプル数 |
| `min_samples_leaf` | 葉ノードに必要な最小サンプル数 |

`n_estimators` を増やすと性能が安定しやすくなりますが、計算時間は長くなります。  
`max_depth` が大きいほど複雑な木を作れますが、大きすぎると過学習しやすくなります。

### 5-8. Random Forestモデルを作る

## 3. 線形回帰

線形回帰は、目的変数を特徴量の足し算で表すモデルです。

```
予測値 = 切片 + 係数1 × 特徴量1 + 係数2 × 特徴量2 + ...
```

特徴量と目的変数の関係を、なるべくよく説明する直線または平面を探すイメージです。

### 3-1. モデルを学習する


````python
lr_reg = LinearRegression()
lr_reg.fit(X_train_reg, y_train_reg)

print("学習完了")
````

### 3-2. 予測する


````python
y_pred_reg = lr_reg.predict(X_test_reg)

result_reg = pd.DataFrame({
    "actual": y_test_reg.values,
    "predicted": y_pred_reg,
})

result_reg.head(10)
````

### 3-3. 回帰の評価指標

回帰では、予測値と実際の値の「ずれ」を評価します。

| 指標 | 意味 | 見方 |
| ---- | ---- | ---- |
| MAE | 絶対誤差の平均 | 小さいほどよい |
| RMSE | 二乗誤差の平均の平方根 | 小さいほどよい、大きな誤差に厳しい |
| R2 | 決定係数 | 1に近いほどよい |


````python
mae = mean_absolute_error(y_test_reg, y_pred_reg)
rmse = np.sqrt(mean_squared_error(y_test_reg, y_pred_reg))
r2 = r2_score(y_test_reg, y_pred_reg)

print(f"MAE : {mae:.3f}")
print(f"RMSE: {rmse:.3f}")
print(f"R2  : {r2:.3f}")
````

### 3-4. 実測値と予測値を可視化する


````python
plt.figure(figsize=(6, 6))
plt.scatter(y_test_reg, y_pred_reg, alpha=0.8)
plt.plot(
    [y_test_reg.min(), y_test_reg.max()],
    [y_test_reg.min(), y_test_reg.max()],
    color="red",
    linestyle="--",
    label="完全に一致する線",
)
plt.xlabel("実際のチップ金額")
plt.ylabel("予測したチップ金額")
plt.title("回帰モデルの予測結果")
plt.legend()
plt.tight_layout()
plt.show()
````

````python
X_train_cls, X_test_cls, y_train_cls, y_test_cls = train_test_split(
    X_cls,
    y_cls,
    test_size=0.2,
    random_state=42,
    stratify=y_cls,
)

print(f"訓練データ: {X_train_cls.shape[0]} 件")
print(f"テストデータ: {X_test_cls.shape[0]} 件")
print("訓練データのクラス比率:")
print(y_train_cls.value_counts(normalize=True))
print("テストデータのクラス比率:")
print(y_test_cls.value_counts(normalize=True))
````

---

## 6. Logistic回帰

Logistic回帰は名前に「回帰」と入っていますが、分類モデルです。線形のスコアを作り、それを確率に変換してクラスを予測します。

特徴量のスケールが大きく違うと学習が不安定になることがあるため、今回は `StandardScaler` と `LogisticRegression` を `Pipeline` でつなぎます。


````python
logistic_model = Pipeline([
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(max_iter=1000, random_state=42)),
])

logistic_model.fit(X_train_cls, y_train_cls)
y_pred_cls = logistic_model.predict(X_test_cls)

print("正解率:", accuracy_score(y_test_cls, y_pred_cls))
print(classification_report(y_test_cls, y_pred_cls, target_names=cancer.target_names))
````

### 6-1. 混同行列


````python
cm = confusion_matrix(y_test_cls, y_pred_cls)

ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=cancer.target_names,
).plot(cmap="Blues", colorbar=False)

plt.title("Logistic回帰の混同行列")
plt.tight_layout()
plt.show()
````

混同行列では、どのクラスをどのクラスと間違えたかを確認できます。医療・安全・不正検知のような場面では、単なる正解率だけではなく、どちらの間違いが重大かを考える必要があります。

## 7. Random Forest 分類と比較する

同じ分類問題を Random Forest でも解いてみます。


````python
rf_cls = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
)

rf_cls.fit(X_train_cls, y_train_cls)
y_pred_rf_cls = rf_cls.predict(X_test_cls)

print("正解率:", accuracy_score(y_test_cls, y_pred_rf_cls))
print(classification_report(y_test_cls, y_pred_rf_cls, target_names=cancer.target_names))
````

### 7-1. 複数モデルを表で比較する


````python
cls_results = []

classification_models = {
    "Logistic Regression": logistic_model,
    "Random Forest Classifier": rf_cls,
}

for name, model in classification_models.items():
    model.fit(X_train_cls, y_train_cls)
    y_pred = model.predict(X_test_cls)
    acc = accuracy_score(y_test_cls, y_pred)

    cls_results.append({
        "model": name,
        "accuracy": acc,
    })

cls_compare = pd.DataFrame(cls_results)
cls_compare
````

## 8. 回帰と分類をまとめて比較する

ここまでの内容を整理します。

| 観点 | 回帰 | 分類 |
| ---- | ---- | ---- |
| 目的変数 | 連続値 | カテゴリ、ラベル |
| 予測結果 | 3.25, 108.7, 0.82 など | 0/1, A/B/C, 良性/悪性 |
| よく使うモデル | LinearRegression, RandomForestRegressor | LogisticRegression, RandomForestClassifier |
| 評価指標 | MAE, RMSE, R2 | accuracy, precision, recall, F1 |
| 可視化 | 実測値 vs 予測値 | 混同行列、分類境界、ROC曲線 |

scikit-learn ではモデル名の末尾を見ると、回帰用か分類用か分かることがあります。

| モデル名 | 用途 |
| ---- | ---- |
| `RandomForestRegressor` | 回帰 |
| `RandomForestClassifier` | 分類 |
| `LinearRegression` | 回帰 |
| `LogisticRegression` | 分類 |

> [!NOTE]
> `LogisticRegression` は分類モデルです。名前だけ見ると回帰のように感じますが、実際にはクラスに属する確率を出して分類します。
