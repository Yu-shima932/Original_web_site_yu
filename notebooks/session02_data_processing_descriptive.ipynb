## 0. 表形式データの基本（CSVとExcel）

データサイエンスで扱うデータの多くは、行列で構成される「表形式（Tabular Data）」です。Pythonで処理する前に、代表的な2つの形式の違いを理解しておきましょう。

### 0.1 CSV (Comma Separated Values)
- **特徴**: データを「カンマ（,）」で区切っただけのシンプルなテキストファイル。
- **メリット**: ファイルサイズが小さく、どんなソフトでも開ける（互換性が非常に高い）。
- **デメリット**: セルの色、数式、複数のシートなどを保存することはできない。

### 0.2 Excel (.xlsx)
- **特徴**: Microsoft Excelで使われるバイナリ形式のファイル。
- **メリット**: 複数のシート、グラフ、書式（色や太字）、数式を保持できる。
- **デメリット**: ファイルサイズが大きくなりがちで、読み込みに専用のライブラリ（Pythonでは `openpyxl` など）が必要。

### 0.3 行（Row）と列（Column）
データフレーム（Pandasで扱う表）では、以下のように呼びます。
- **列 (Column / Variable)**: 縦方向。「項目」を表す（例：身長、体重、疾患名）。
- **行 (Row / Observation)**: 横方向。「一件ごとのデータ」を表す（例：Aさんのデータ、Bさんのデータ）。

## 3. 代表値と数値データの性質（パラメトリック・ノンパラメトリック）

数値データを1つの数字で代表させる場合、**平均値（Mean）** と **中央値（Median）** のどちらを使うかが非常に重要です。

### 3.1 平均値、SD、95%信頼区間
データが正規分布に従う（パラメトリックな）場合、以下の3つの指標がセットでよく使われます。

1. **平均値 (Mean)**: データの合計を個数で割った値。
2. **中央値 (Median)**: データを大きさ順に並べたときにちょうど真ん中にくる値。外れ値の影響を受けにくいのが特徴です。
3. **標準偏差 (SD: Standard Deviation)**: データの「ばらつき」の平均的な大きさ。値が大きければデータが広く分散していることを意味します。
4. **信頼区間 (CI: Confidence Interval)**: 推定した統計量（平均値や中央値など）が、母集団においてどの程度の範囲に収まるかを確率的に示したものです。データの精度や「確からしさ」を表します。

### 3.2 平均値と中央値：いつ一致し、いつズレるのか？

データの「分布」の形によって、これら2つの指標の関係が変わります。これが統計手法（パラメトリック vs ノンパラメトリック）を選ぶ大きなヒントになります。

#### 1. 一致する場合（左右対称）
- データが綺麗な「左右対称（正規分布など）」であれば、平均値と中央値はほぼ一致します。
- **例**: 成人の身長データなど。この場合は**パラメトリック手法**が適しています。


````python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 1. 左右対称なデータ（正規分布）の生成
np.random.seed(42)
normal_data = np.random.normal(loc=170, scale=5, size=1000)

plt.figure(figsize=(8, 4))
sns.histplot(normal_data, kde=True, color="skyblue")
plt.axvline(np.mean(normal_data), color="red", linestyle="--", label=f"Mean: {np.mean(normal_data):.1f}")
plt.axvline(np.median(normal_data), color="blue", linestyle="-", label=f"Median: {np.median(normal_data):.1f}")
plt.title("Symmetric Distribution (Normal)")
plt.legend()
plt.show()
````

#### 2. 一致しない場合（非対称・外れ値あり）
- データがどちらかに偏っている（歪んでいる）場合や、極端な**外れ値**がある場合、平均値は外れ値に引っ張られて大きくズレますが、中央値は安定しています。
- **例**: 日本の世帯所得（一部の超高所得者が平均値を押し上げる）、入院日数（一部の長期入院患者が平均値を押し上げる）。
- このように分布が偏っている場合は、**ノンパラメトリック手法**（中央値ベース）で考えるのが妥当です。


````python
# 2. 右に裾が長いデータ（偏りのある分布）の生成
skewed_data = np.random.lognormal(mean=2, sigma=0.8, size=1000)

plt.figure(figsize=(8, 4))
sns.histplot(skewed_data, kde=True, color="orange")
plt.axvline(np.mean(skewed_data), color="red", linestyle="--", label=f"Mean: {np.mean(skewed_data):.1f}")
plt.axvline(np.median(skewed_data), color="blue", linestyle="-", label=f"Median: {np.median(skewed_data):.1f}")
plt.title("Skewed Distribution (Asymmetric)")
plt.legend()
plt.show()
````

### 3.2 正規分布（Normal Distribution）とは？
左右対称の「釣り鐘型（ベルカーブ）」の分布です。
- 平均値を中心に、±1SDの範囲に約68%、±2SDの範囲に約95%のデータが含まれるという性質があります。

### 3.3 なぜ「正規分布かどうか」が重要なのか？
データが正規分布に従うかどうかで、**使用できる統計手法（検定）が全く異なる**からです。

| データの性質   | 手法（パラメトリック）                   | 手法（ノンパラメトリック）                   |
| :------------- | :--------------------------------------- | :------------------------------------------- |
| **分布の仮定** | 正規分布に従う                           | 正規分布に従わない（または不明）             |
| **要約指標**   | **平均値 ± SD**                          | **中央値 (四分位範囲)**                      |
| **主な検定**   | t検定、ANOVA（等分散性が必要）           | ウィルコクソン検定、クラスカル・ウォリス検定 |
| **特徴**       | **統計的検定力（差を見つける力）が高い** | 外れ値に強く、頑健（ロバスト）               |

> 💡 **なぜ使い分けるのか？**:
> 1. **検定力の違い**: 正規分布に従うデータにノンパラメトリックな手法を使うと、本当は差があるのに「差がない」と判定してしまう（検定力が下がる）リスクがあります。
> 2. **妥当性の維持**: 逆に、分布が偏っているデータにパラメトリックな手法を使うと、偽の結果（誤った結論）を導く可能性があります。

### 3.4 使い分けの基準（まとめ）
- **パラメトリックデータ**: データの分布が正規分布に従っていると判断できる場合。**平均値やSDを用いて報告**します。
    - 表記例: `170.5 ± 5.2` (Mean ± SD) や `170.5 [95% CI: 168.2 - 172.8]` (CI)
- **ノンパラメトリックデータ**: データが正規分布に従わない（例：所得、入院日数など）、または外れ値の影響が大きすぎる場合。**中央値を用いて報告**します。実務的には、データのばらつきを示す**四分位範囲 (IQR)** や、推定の精度を示す **95% 信頼区間 (CI)** を併記するのが一般的です。
    - 表記例: `50.0 [25.0 - 75.0]` (IQR) や `50.0 [95% CI: 45.0 - 55.0]` (CI)


````python
# 平均値
print(f"平均値: {df['total_bill'].mean():.2f}")

# 標準偏差 (SD)
print(f"標準偏差 (SD): {df['total_bill'].std():.2f}")

# 中央値
print(f"中央値: {df['total_bill'].median():.2f}")

# 四分位範囲 (IQR: Interquartile Range)
Q1 = df['total_bill'].quantile(0.25)
Q3 = df['total_bill'].quantile(0.75)
IQR = Q3 - Q1
print(f"第1四分位点 (Q1 / 25%): {Q1:.2f}")
print(f"第3四分位点 (Q3 / 75%): {Q3:.2f}")
print(f"IQR (Q3 - Q1): {IQR:.2f}")
print(f"ノンパラメトリック表記: {df['total_bill'].median():.2f} [{Q1:.2f} - {Q3:.2f}]")
````

## 4. df.describe() の出力とその意味

データ全体を俯瞰するために最も便利なコマンドが `df.describe()` です。


````python
# 数値データの要約統計量を表示
df.describe()
````

### 各項目の意味
- **count**: データの数（欠損値を除いた有効な数）
- **mean**: 平均値
- **std**: 標準偏差（データの「ばらつき」の大きさ）
- **min**: 最小値
- **25%**: 第1四分位点（データを小さい順に並べて下から25%の位置）
- **50%**: 中央値（第2四分位点）
- **75%**: 第3四分位点（下から75%の位置）
- **max**: 最大値

> 💡 **ポイント**: `mean`（平均）と `50%`（中央値）の差が大きい場合、そのデータは左右対称ではなく、どちらかに長く尾を引いた分布（歪んだ分布）である可能性が高いです。

### どのグラフを使えばよいか？

グラフの選択は、「**何を・何と比べて・何のために見せるか**」によって決まります。以下を目安にしてください。

| 目的                         | 変数の組み合わせ | 推奨グラフ                         |
| :--------------------------- | :--------------- | :--------------------------------- |
| **分布の形を見る**           | 数値1変数        | ヒストグラム (`histplot`)          |
| **グループ間を比較する**     | カテゴリ × 数値  | 箱ひげ図 (`boxplot`)               |
| **2変数の相関を見る**        | 数値 × 数値      | 散布図 (`scatterplot` / `regplot`) |
| **時系列・順序の推移を見る** | 順序 × 数値      | ラインプロット (`lineplot`)        |
| **割合や構成比を見る**       | カテゴリ → 数値  | 棒グラフ (`barplot`)               |

### 5.1 ヒストグラム（分布の確認）
データが正規分布しているか、どこに塊があるかを確認します。


````python
plt.figure(figsize=(8, 5))
sns.histplot(df['total_bill'], kde=True) # kde=True で密度曲線を表示
plt.title('Distribution of Total Bill')
plt.show()
````

### 5.2 箱ひげ図（Boxplot：要約と外れ値）
中央値、四分位範囲、そして「外れ値（点として表示される）」を一目で確認できます。


````python
plt.figure(figsize=(8, 5))
sns.boxplot(x='day', y='total_bill', data=df)
plt.title('Total Bill by Day')
plt.show()
````

### 5.3 散布図と回帰直線（Scatter Plot & Trend Line）
2つの数値データの関連性を確認します。


````python
plt.figure(figsize=(8, 5))
# regplot は散布図に回帰直線（ライン）を重ねて表示します
sns.regplot(x='total_bill', y='tip', data=df, line_kws={'color': 'darkred'})
plt.title('Total Bill vs Tip with Regression Line')
plt.show()
````

### 5.4 ラインプロット（Line Plot：推移や順序）
時系列や、特定の変化（dose-responseなど）を見る時に使用します。


````python
# 日ごとの合計金額の平均推移（デモ用の集計）
daily_avg = df.groupby('day', observed=False)['total_bill'].mean().reset_index()

plt.figure(figsize=(8, 5))
sns.lineplot(x='day', y='total_bill', data=daily_avg, marker='o', linewidth=2.5)
plt.title('Average Total Bill Trend by Day')
plt.show()
````

## 5.5 groupby：グループ別の集計

`groupby` は、カテゴリ変数でデータを**グループに分けて集計**するための最重要操作です。「種類ごとの平均を比べたい」「曜日ごとの合計を出したい」といった場面で必須となります。


````python
# 基本構文: df.groupby('グループ列名')['集計する列名'].集計関数()

# 曜日（day）ごとのチップ（tip）の平均値
day_mean = df.groupby('day', observed=False)['tip'].mean()
print(day_mean)

# 複数の統計量を一括で確認
day_stats = df.groupby('day', observed=False)['total_bill'].agg(['mean', 'median', 'std', 'count'])
print(day_stats)
````

> 💡 **ポイント**: `observed=False` は、カテゴリ型の列を `groupby` する際に、データが存在しないカテゴリも結果に表示するための引数です。Pandas 2.0以降では省略すると警告が表示されます。
